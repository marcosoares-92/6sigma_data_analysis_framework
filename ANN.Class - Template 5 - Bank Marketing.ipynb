Os algoritmos inspirados na estrutura e funcionamento do cérebro são geralmente chamados de Redes Neurais Artificiais (“Artificial Neural Networks”, ANN). Uma subclasse das redes neurais, por sua vez, é o Deep Learning (“aprendizado profundo”, em tradução livre). 

As redes neurais são uma classe de algoritmos que tenta encontrar padrões para os dados. Elas podem ser utilizadas tanto em problemas de regressão quanto de classificação. Na realidade, trata-se de um enorme campo do Machine Learning formado por centenas de algoritmos e por suas variações, que permitem a sua adaptação à solução de toda sorte de problemas.

As camadas atuam de forma semelhante aos neurônios biológicos: as saídas de uma camada servem como entrada da camada seguinte. Dentre as camadas, podemos distinguir a camada de entrada (“input layer”), as camadas ocultas (“hidden layers”), e uma camada de saída (“output layer”). Comumente, as redes são completamente conectadas, o que significa que existe uma conexão entre cada neurônio de uma camada e cada neurônio da camada seguinte. Embora a conexão total não seja um requisito, é o caso típico nas ANNs.

os princípios das redes neurais artificiais (ANNs) são gerais, de modo que esta mesma ferramenta poderá ser aplicada aos mais diversos processos de tomada de decisão e de análise de propensão ou de risco, ou de regressão para predição de valores de interesse. 

###### Para a classificação, as ANNs (assim como a regressão logística) fornecem se um determinado evento é mais propenso a ocorrer ou a não ocorrer; e qual a probabilidade deste evento ocorrer. Porém, enquanto a regressão logística analisava apenas saídas binárias (evento ocorre ou não ocorre), as ANNs podem dividir um evento de saída em múltiplas categorias, sendo mais adequadas para as situações nas quais a saída categórica pode ser descrita por vários eventos possíveis e contáveis. 

Não bastasse, os algoritmos ANN são mais poderosos, podendo alcançar maior precisão, e são os melhores algoritmos disponíveis para problemas de processamento de imagens. Contudo, o uso de redes neurais implica maior custo computacional, fazendo com que o algoritmo leve mais tempo e consuma mais memória. 

Além de apresentar convergência mais rápida e menor custo computacional, a regressão logística apresenta fundamentação estatística bastante sólida e fornece diretamente um valor de probabilidade, facilitando a criação de escores de risco. Sendo assim, embora possam ser usadas para situações iguais ou próximas, são duas ferramentas que se complementam. Exemplos disso: podemos usar a regressão logística em um primeiro momento para obter mais rapidamente o gráfico Shap; eliminar variáveis; e só então usar as redes neurais; ou podemos tomar um subconjunto binário das redes neurais e aplicar a regressão logística a ele para obter a probabilidade relacionada aos eventos desse subconjunto (para construir um escore de risco, por exemplo).

Alternativamente ao uso de regressão logística, outro algoritmo ML mais simples e que fornece explicações simples e confiáveis, podendo ser usado em conjunto para explicar os resultados das redes neurais é o Decision Tree. Como sugerido pelo nome, o Decision Tree fornece os resultados e respectivas probabilidades e critérios de corte na forma de árvore de decisão, sendo bastante útil para interpretação dos resultados de um modelo mais complexo.

##### Problema de classificação

Um problema de classificação existe sempre que a variável de saída é uma categoria, como “vermelho” ou “azul”, “doente”, ou “sadio”. Quando existem apenas duas classificações possíveis, a regressão logística é uma ferramenta adequada, inclusive pelo rigor estatístico com que é definida. Porém, problemas que podem resultar em mais de duas classificações (num exemplo trivial: “vermelho”, “azul”, “verde”, “laranja”) exigem ferramentas mais poderosas, como as redes neurais classificatórias.

Assim, problemas clássicos de grande interesse prático que podem ser desenvolvidos em termos de classificação são os que envolvem a realização de recomendações: comprar ou não comprar; pagar ou não pagar; classificar como “verde”, “amarelo”, ou como azul”. Os problemas de escolha e recomendação são tipicamente problemas de classificação; enquanto que prever um valor numérico a partir de dados históricos é tipicamente um problema de regressão.

Como dito acima, a regressão logística não pode ser usada em complemento às redes neurais para explicar a saída de um modelo com múltiplas classificações. Neste caso, o algoritmo mais adequado para complementar e explicar a saída das ANN é a Decision Tree (a qual também pode ser aplicada ao caso binário, no qual a árvore construída apresentará apenas 2 ramos de decisão).

##### Multi-Layer Perceptrons (MLP) para classificação

Redes de perceptrons, as redes neurais, são perceptrons multi-camadas (“multi-layer”). A figura a seguir foi publicada na Dissertação de Mestrado de Luca Longo, na University of Dublin (Irlanda). Nela, o autor ilustra o uso de uma rede neural capaz de analisar diversas características de um paciente e prever se ele terá determinados problemas cardíacos. Veja na figura que temos 4 classificações de saída possíveis, o que impossibilita que uma análise completa seja feita por um modelo de saída binária, como a regressão logística.



https://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.799.8752&rep=rep1&type=pdf 

Esta rede neural apresenta 6 variáveis de entrada (os “Attributes”, obtidos por análises de proteínas do corpo), as quais são usadas como input da camada de entrada. Note que a camada de entrada (“input layer”) é mostrada em verde; as camadas ocultas (“hidden layers”) estão em azul; e a camada de saída (“output layer”) é mostrada em vermelho, havendo 4 classes/classificações de saída possíveis (as “target classes”). 

Assim, o algoritmo pode ser descrito como uma “rede de operações matemáticas”: um conjunto de uma ou mais variáveis de entrada passa pela rede de operações matemáticas dos neurônios, resultando em uma ou mais variáveis de saída. As redes neurais não são limitadas a problemas lineares, sendo capazes de representar limites relacionados a decisões complexas.

Assim, as ANN ou “multi-layer perceptrons” (MLP) são algoritmos supervisionados que aprendem uma função que correlaciona a saída a um padrão de entrada por meio do treinamento em um dataset. Ao contrário do perceptron, que possuía saída única, as redes neurais podem ter múltiplas saídas, correspondentes a cada um dos neurônios na camada de saída (“output layer”). O MLP pode aprender uma função não-linear (ativação) tanto para o caso de classificação quanto para o de regressão.

As redes neurais classificatórias se diferenciam da regressão logística, portanto, no sentido de admitirem uma ou mais camadas não-lineares (as “hidden layers”). As vantagens de uso das redes neurais incluem: a capacidade de aprendizado de modelos não-lineares; 

###### e a capacidade de aprendizado de modelos em tempo real (“on-line learning”). No caso da sklearn, isto é feito por meio do comando partial_fit (ao invés do fit).

Por fim, vamos relembrar algumas desvantagens relacionadas ao uso do MLP que vimos no Módulo 2:

- As hidden layers apresentam funções de perda/erro (“loss function”) não-convexas nas quais pode existir mais de um mínimo local. Portanto, diferentes inicializações aleatórias dos pesos podem resultar em diferentes precisões de validação.
- O algoritmo MLP requer a sintonia/ajuste/definição de vários parâmetros, como o número de neurônios nas hidden layers, número de camadas; número de epochs (total de vezes que todos os dados do dataset serão utilizados durante o processo de backpropagation); e número de iterações. Alguns parâmetros podem ser mantidos de acordo com o padrão (default), mas é muito comum precisar ajustar manualmente o número de iterações para possibilitar a convergência. 
- No Módulo 2, vimos que isto é feito informando-se um valor max_iter nos argumentos da função de chamamento das redes neurais.
- Assim como ocorre com os perceptrons, as redes neurais classificadoras são extremamente sensíveis aos fatores de escala. Portanto, é altamente recomendável proceder à normalização dos dados antes da análise.
- Neste caso, o algoritmo da sklearn será o MLPClassifier (sklearn.neural_network.MLPClassifier). As informações sobre ele se encontram em: https://scikit-learn.org/stable/modules/neural_networks_supervised.html?highlight=back e em https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html#sklearn.neural_network.MLPClassifier.

- Definir o solver utilizado. Mantenha o ‘adam’ como padrão, ou selecione um outro, de acordo com a documentação da scikit-learn.
- Definir o número de camadas ocultas (hidden layers) e de neurônios por camada oculta. Isto é informado em hidden_layer_sizes. Por exemplo: hidden_layer_sizes=100 indica que a rede será formada por uma única hidden layer contendo 100 neurônios. Caso deseje testar mais camadas ocultas, deve ser fornecida uma lista. # 4. Avaliação de desempenho do modelo 

Como voltamos ao problema de classificação, usaremos as métricas aplicadas à regressão logística, e avaliaremos a matriz de confusão (que agora pode possuir mais de 4 quadrantes, cada um correspondente a cada uma das classificações possíveis).

A exatidão/precisão (“accuracy”) do modelo é definida a partir dos resultados do modelo como a (fração de predições corretas) = (número de predições corretas)/(número total de dados usados para teste). Por sua vez, o número de dados é a quantidade de linhas (entradas) do conjunto de dados (lembre-se: as colunas são as variáveis). Podemos definir a precisão tanto em termos do desempenho do modelo com relação à predição para os dados de teste, quanto com relação ao ajuste dos dados de treino. Alguns conceitos importantes:

- Positivo verdadeiro: o modelo prevê corretamente um resultado positivo (na regressão logística: prevê saída 1 e o resultado era 1);
- Negativo verdadeiro: o modelo prevê corretamente um resultado negativo (na regressão logística: prevê saída 0 e o resultado era 0);
- Falso positivo (erro tipo 1): o modelo prevê incorretamente o resultado positivo para um dado cuja classe é negativa (na regressão logística: prevê saída 1, mas o resultado era 0);
- Falso negativo (erro tipo 2): o modelo prevê incorretamente o resultado negativo para um dado cuja classe é positiva (na regressão logística: prevê saída 0, mas o resultado era 1).

Podemos gerar um relatório das métricas classificatórias (“classification report”), as quais auxiliam a avaliação dos erros tipos 1 e 2. Em primeiro lugar, repare que “accuracy” é traduzido como precisão ou exatidão (a palavra “acurácia” é neologismo não utilizado na área de instrumentação). A mesma tradução é utilizada para a palavra “precision”. Como as métricas aqui são ligeiramente diferentes, vamos manter o termo em inglês, facilitando a leitura do relatório:

1. Accuracy: como já vimos, é a relação entre o número de previsões corretas e o número total de previsões.

Accuracy=((Número de previsões corretas))/((Número total de previsões) )  (1)


2. Recall: se refere à tentativa de responder à pergunta: “Qual é a proporção de positivos verdadeiros (PV) que foi identificada corretamente?”. Em outras palavras, é a relação entre o número PV e o total de dados que eram efetivamente positivos, i.e., a soma entre PV e os falsos negativos (FN, dados positivos (1) que foram erroneamente considerados negativos, sendo atribuídos à classe zero).

Recall=PV/((PV+FN) )  (2)

3. Precision: se refere à tentativa de responder à pergunta: “Qual é a proporção de identificações positivas que estava efetivamente correta?”. Em outras palavras, é a relação entre o número PV e o total de dados classificados como positivos, i.e., a soma entre PV e os falsos positivos (FP, dados negativos (0) que foram erroneamente considerados positivos, sendo atribuídos à classe 1).

Precision=PV/((PV+FP) )  (3)

4. F1-score: mais uma medida da precisão do modelo. Consiste na média harmônica entre “precision” e “recall”. Sendo assim, pode ter um valor máximo igual a 1 (“precision” e “recall” perfeitos), e um valor mínimo zero. De forma simples, é uma medida da precisão e robustez do modelo.

F1=2/((1/Precision+1/Recall) )=(2∙(Precision)∙(Recall))/((Precision+Recall) )  (4)

Uma última métrica é o “AUC-ROC score”. Em um problema genérico de classificação, esta medida nos diz o quanto o modelo é capaz de distinguir as classes.



https://towardsdatascience.com/how-to-evaluate-your-machine-learning-models-with-python-code-5f8d2d8d945b

```python
# Use score method to get accuracy of model
accuracy_train = ann_class.score(x_train, y_train)
print(accuracy_train)

accuracy_test = ann_class.score(x_test, y_test)
print(accuracy_test)

#Full reports
report_train = classification_report(y_train, y_pred_train)
print(report_train)

report_test = classification_report(y_test, y_pred_test)
print(report_test)
```

# 5. Matriz de Confusão (“Confusion Matrix”) 

Vamos agora analisar o desempenho sob a ótica da matriz de confusão (um dataset de dígitos). A matriz de confusão é uma tabela comumente utilizada para descrever o desempenho de um modelo de classificação (ou “classificador”) para um conjunto de dados de teste para os quais se conhece os valores verdadeiros.

A matriz de confusão é dividida em vários quadrantes, sendo que cada quadrante representa uma possível classificação. No caso da regressão logística (modelos binários), existem apenas duas classificações possíveis: 1 ou zero. 

Quanto mais elementos na diagonal principal, mais preciso é o modelo de classificação. Isso porque todos os acertos do modelo (previu a classificação que efetivamente se verificou) se encontram sobre a diagonal principal. Repare que isso se observa porque o eixo vertical é organizado de baixo para cima (ou seja, ele cresce em direção à parte inferior da matriz. Caso ele crescesse de cima para baixo, a diagonal secundária é que seria utilizada).

Se existem n possíveis classificações, teremos n (horizontal) x n (vertical) = n² quadrantes. Como a regressão logística possui apenas n = 2 classificações (1 ou zero), a matriz de confusão será dividida em 2 x 2 = 2² = 4 quadrantes, cada um representando uma das possíveis situações.

Em cada quadrante, a matriz mostra a quantidade de elementos para os quais se verificou cada uma destas situações, e impõe uma escala de cores (mapa de calor) para facilitar a visualização dos dados. Ilustrando para a classificação binária:

1.	Modelo prevê o resultado zero, e se verifica o resultado zero (previsão correta);
2.	Modelo prevê o resultado zero, mas se verifica o resultado 1 (erro do modelo);
3.	Modelo prevê o resultado 1, e se verifica o resultado 1 (previsão correta);
4.	Modelo prevê o resultado 1, mas se verifica o resultado zero (erro do modelo).

Em cada quadrante, a matriz mostra a quantidade de elementos para os quais se verificou cada uma destas situações, e impõe uma escala de cores (mapa de calor) para facilitar a visualização dos dados.



Assim, vamos construir as matrizes de confusão para ambos os conjuntos de dados, treino e teste. Ao construir a matriz de confusão, podemos mostrar os números das quantidades de dados em cada quadrante ou não. Para mostrar os números, mantenha annot=True. Caso não deseje vê-los, mude para annot=False. Quando muitas classes estão presentes, é interessante ocultar os números para que o gráfico não fique poluído. Repare que a interpretação de cada quadrante continua a mesma: no eixo vertical a classe real, no horizontal a classe predita.

```python
cm_train = metrics.confusion_matrix(y_train, y_pred_train)

plt.figure(figsize=(9,9))
sns.heatmap(cm_train, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
#annot = True: shows the number corresponding to each square
#annot = False: do not show the number
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score (Train): {0}'.format(accuracy_train)
plt.title(all_sample_title, size = 15);

cm_test = metrics.confusion_matrix(y_test, y_pred_test)

plt.figure(figsize=(9,9))
sns.heatmap(cm_test, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
#annot = True: shows the number corresponding to each square
#annot = False: do not show the number
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score (Test): {0}'.format(accuracy_test)
plt.title(all_sample_title, size = 15);
```

Existem diversas interações entre elas, sugerindo que cada combinação de características que define uma paciente seja analisada em particular e de forma holística.

##### Assim, após reduzir o total de variáveis de 19 para 10, chegamos à seguinte ordem de priorização de clientes pela equipe de telemarketing, de acordo com os fatores que resultam em maior propensão a tomar o crédito. Note que alguns destes fatores são macroeconômicos e sociais, de caráter geral, afetando a propensão do conjunto de clientes como um todo a tomar o crédito.

1. Menor taxa de variação trimestral do índice de ocupação (índice de pessoas empregadas);
2. Maior taxa de juros (Euribor);
3. Maior índice de preços ao consumidor (indicador de inflação mensal);
4. Menores índices de confiança do consumidor (indicador mensal)
5. Clientes que foram contatados há menos tempo, na última campanha;
6. Clientes que receberam menos contatos na presente campanha;
7. Clientes que já estão inadimplentes.

- Além disso, os clientes devem ser preferencialmente contatados por telefone (e não celular); e clientes que receberam o último contato há muito tempo ou não foram contatados na última campanha não devem ser priorizados, sendo que o mesmo vale para clientes que já receberam muitos contatos na campanha presente e ainda não contrataram o empréstimo. Fatores como o resultado da campanha anterior e o dia da semana em que será feito o contato podem tanto aumentar quanto reduzir a propensão do cliente a tomar empréstimo, e devem ser analisados individualmente antes do contato.

- Por fim, embora os clientes já inadimplentes tenham maior propensão a tomar o crédito, emprestar a eles implica incorrer em risco maior de não receber no futuro, de modo que taxas de juros mais altas são recomendadas (como discutido no problema de regressão logística, no qual calculamos o escore de risco com base na probabilidade de cada cliente pagar ou não as dívidas).
